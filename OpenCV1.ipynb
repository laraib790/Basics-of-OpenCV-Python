# Image threshold

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/gdrive/My Drive/Colab Notebooks/test.jpg')
cv2_imshow(img)

grey1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(grey1.shape)
cv2_imshow(grey1)

ret,thresh1 = cv2.threshold(grey1,130,255,cv2.THRESH_BINARY)
print(thresh1.shape)
cv2_imshow(thresh1)

ret,thresh1 = cv2.threshold(grey1,130,255,cv2.THRESH_BINARY_INV)
print(thresh1.shape)
cv2_imshow(thresh1)
```

# Color Detection
RGB method


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/gdrive/My Drive/Colab Notebooks/all_color.jpg')
print(img.shape)
cv2_imshow(img)

Blue_Channel = img[:,:,0]
Green_Channel = img[:,:,1]
Red_Channel = img[:,:,2]

print("Blue Chanel")
cv2_imshow(Blue_Channel)

grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print("Printing Grey image")
print(grey.shape)
cv2_imshow(grey)

Blue = cv2.subtract(Blue_Channel,grey)
print("Subtracted Image")
cv2_imshow(Blue)

ret,thresh1 = cv2.threshold(Blue,80,255,cv2.THRESH_BINARY)
print("Printing threshlod image")
print(thresh1.shape)
cv2_imshow(thresh1)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/gdrive/My Drive/Colab Notebooks/all_color.jpg',1)
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

#define range of blue color based on hsv value we got

Blue_low = np.array([110,50,50])
Blue_high = np.array([130,255,255])

#thresholding to get only blue color
mask = cv2.inRange(hsv,Blue_low,Blue_high)

cv2_imshow(img)
cv2_imshow(mask)
```

# Drawing on an image

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

img = np.zeros((512,512,3),np.uint8)
cv.line(img,(0,0),(512,512),(256,0,0),5)
cv2_imshow(img)

```

```python
cv.rectangle(img,(0,0),(256,256),(0,0,255),10)
cv2_imshow(img)
```

```python
cv.circle(img,(256,256),100,(0,255,255),8)
cv2_imshow(img)
```

```python
font = cv.FONT_HERSHEY_COMPLEX
cv.putText(img,'Hello',(10,500),font,4,(255,0,255),2)
cv2_imshow(img)
```
